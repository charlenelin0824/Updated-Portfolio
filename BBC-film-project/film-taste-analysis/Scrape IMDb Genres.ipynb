```python
import pandas as pd
import numpy as np

```

```python
# read csv "OMDb-cleaned" into a pandas dataframe
df = pd.read_csv("OMDb_cleaned.csv")
df
```

```python
# Create a new df called "film-IMDb-ID" that contains only the columns "Title" and "imdbID"
df_imdb_id = df[["Title", "imdbID"]]
df_imdb_id
```

```python
# read tsv "title.basics.tsv" and show the first 5 rows
df_title_basics = pd.read_csv("title.basics.tsv", sep="\t")
df_title_basics.head()
```

```python
df_list = []
for index, row in df_imdb_id.iterrows():
    imdb_id = row["imdbID"]
    film_info = df_title_basics[df_title_basics["tconst"] == imdb_id]
    df_list.append(film_info)
df_imdb_id_basics = pd.concat(df_list)
```

```python
# prepare to scrape a page
import requests
from bs4 import BeautifulSoup
```

```python
# pretend to be a browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}
```

```python
# scrape the page for all the plot key words
def get_film_info(imdb_id):
    url = f"https://www.imdb.com/title/{imdb_id}/keywords/"
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")
    plot_keywords= soup.find_all("a", class_="ipc-metadata-list-summary-item__t")
    plot_keywords_list = []
    for keyword in plot_keywords:
        plot_keywords_list.append(keyword.text)
    return plot_keywords_list
```

```python
for id in df_imdb_id_basics["tconst"]:
    print(get_film_info(id))
```

```python
# count the number of times each keyword appears in the list
from collections import Counter
keyword_list = []
for id in df_imdb_id_basics["tconst"]:
    keyword_list.extend(get_film_info(id))
keyword_count = Counter(keyword_list)
keyword_count
```

```python
#replace items in keyword_list that contains "nudity" with "nudity". For example: replace "female nudity" with "nudity"
keyword_list = [keyword if "nudity" not in keyword else "nudity" for keyword in keyword_list]
# replace items in keyword_list that contains "relationship" with "relationship". For example: replace "father son relationship" with "relationship"
keyword_list = [keyword if "relationship" not in keyword else "relationship" for keyword in keyword_list]
# replace items in keyword_list that contains "sex" with "sex".
keyword_list = [keyword if "sex" not in keyword else "sex" for keyword in keyword_list]
keyword_count = Counter(keyword_list)
keyword_count
```

```python
# turn the list of keywords into a dataframe with the imdbID and keywords list
df_list = []
for id in df_imdb_id_basics["tconst"]:
    keywords = get_film_info(id)
    df_list.append({"imdbID": id, "Title": df_imdb_id_basics['primaryTitle'], "keywords": keywords})
```

```python
df_list
```

```python
# turn df_list into a pandas dataframe
df_keywords = pd.DataFrame(df_list)
df_keywords
```

```python
# drop the Title column
df_keywords = df_keywords.drop(columns=["Title"])
df_keywords
```

```python
# Join df_keywords with df_imdb_id_basics on the imdbID == tconst
df_imdb_id_basics_keywords = df_imdb_id_basics.merge(df_keywords, left_on="tconst", right_on="imdbID")
df_imdb_id_basics_keywords
```
