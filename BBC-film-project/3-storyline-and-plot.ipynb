```python
# read csv into data frame
df = pd.read_csv("OMDb_cleaned.csv")
# show the dataframe with all columns visible
pd.set_option('display.max_columns', None)
df.head(15)
```

```python
# sort by IMDB rating, and show the title as well
IMDb_top_30 = df.sort_values(by='imdbRating', ascending=False).head(30)
IMDb_top_30[['Title', 'imdbRating','Genre']]
```

```python
# split the genre column into multiple columns
IMDb_top_30['Genre'] = IMDb_top_30['Genre'].str.split(', ')
IMDb_top_30 = IMDb_top_30.explode('Genre')

# see IMDb_top_15's genre distribution
IMDb_top_30['Genre'].value_counts()
```

```python
# sort by metacritic rating, and show the title as well
metacritic_top_30 = df.sort_values(by='Metascore', ascending=False).head(30)
metacritic_top_30[['Title', 'Metascore', 'Genre']]
```

```python
# split the genre column into multiple columns
metacritic_top_30['Genre'] = metacritic_top_30['Genre'].str.split(', ')
metacritic_top_30 = metacritic_top_30.explode('Genre')
# see metacritic_top_15's genre distribution
metacritic_top_30['Genre'].value_counts()
```

```python
# sort by Rotten Tomatoes rating, and show the title as well
rotten_tomatoes_top_15 = df.sort_values(by='Rotten Tomatoes', ascending=False).head(20)
rotten_tomatoes_top_15[['Title', 'Rotten Tomatoes']]
```

```python
# import "BBC-film-list.csv" into data frame and make sure the firt row is the header
critic_votes = pd.read_csv("BBC-film-list.csv", header=0)
critic_votes.head(15)
```

```python
critic_us = critic_votes[critic_votes['Critic Country'] == 'US']
critic_us
```

```python
# join critic_us "Movie" with df "Title" columns, and show "Title", "Country", and "Critic Country". Name the new dataframe "df_us_votes"
df_us_votes = pd.merge(df, critic_us, left_on='Title', right_on='Movie', how='inner')[['Title', 'Country', 'Critic Country']]
df_us_votes
```

```python
# do a value count baed on how many time a "Title" appears in the dataframe
df_us_votes['Title'].value_counts().head(20)
```

```python
critic_rest_of_world = critic_votes[critic_votes['Critic Country'] != 'US']
critic_rest_of_world
```

```python
# join critic_rest_of_world "Movie" with df "Title" columns, and show "Title", "Country", and "Critic Country". Name the new dataframe "df_rest_of_world_votes"
df_rest_of_world_votes = pd.merge(df, critic_rest_of_world, left_on='Title', right_on='Movie', how='inner')[['Title', 'Country', 'Critic Country']]
df_rest_of_world_votes
```

```python
# do a value count baed on how many time a "Title" appears in the dataframe
df_rest_of_world_votes['Title'].value_counts().head(20)
```
