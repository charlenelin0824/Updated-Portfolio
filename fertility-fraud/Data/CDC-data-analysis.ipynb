## Dataset 2: 2021 CDC Data - Patient and Cycle Characteristics

```python
# read csv file: 2021_ART__Patient_and_Cycle_Characteristics_20240530.csv in the "Input" folder
df2 = pd.read_csv('./Input/2021_ART__Patient_and_Cycle_Characteristics_20240530.csv')
df2.head()
```

```python
df2 = df2[df2['Question'] == "What was the percentage of cycles in which patients used donor eggs or embryos?"]
df2
```

```python
# remove rows where "Facility Name" is "National" 
df2 = df2[df2['FacilityName'] != 'National']
```

```python
# Create a pivot table
pivot_table = df2.pivot_table(index=['ZipCode', 'ClinicId'], 
                              values=['Cycle_Count', 'Data_Value_num'], 
                              aggfunc={'Cycle_Count': 'median', 'Data_Value_num': 'sum'})
```

```python
#round the numbers to integer
pivot_table['Cycle_Count'] = pivot_table['Cycle_Count'].round(0).astype(int)
# display the pivot table
pivot_table.head()
```

```python
# check the data type of the ZipCode
pivot_table.index.get_level_values(0).dtype
```

```python
# turn the ZipCode into string
pivot_table.index = pivot_table.index.set_levels(pivot_table.index.levels[0].astype(str), level=0)
```

```python
# remove the ".0" in the ZipCode
pivot_table.index = pivot_table.index.set_levels(pivot_table.index.levels[0].str[:-2], level=0)
pivot_table
```

```python
# Calculate cycles of donor eggs/embryos by clinic
pivot_table['Cycles_of_Donor_Eggs_Embryos'] = (pivot_table['Cycle_Count'] * pivot_table['Data_Value_num']) / 100
```

```python
# Sum up the total cycles of donor eggs/embryos by clinic
total_cycles = pivot_table['Cycles_of_Donor_Eggs_Embryos'].sum()
```

```python
total_cycles = round(total_cycles)
print(f"Total cycles of donor eggs/embryos by clinic: {total_cycles}")
```

```python
national_cycles = pivot_table['Cycle_Count'].sum()
print(f"Total cycles in the nation: {national_cycles}")
```

```python
pecentage_of_donor_cycles = (total_cycles / national_cycles) * 100
print(f"Percentage of cycles in which patients used donor eggs or embryos in 2021: {pecentage_of_donor_cycles:.2f}%")
```

## Dataset 4: 2020 CDC Data

```python
# read csv "Preliminary-2020-National-Summary-Clinic-Table.xlsx" in the "Input" folder
df4 = pd.read_excel('./Input/Preliminary-2020-National-Summary-Clinic-Table.xlsx', sheet_name='Clinic Table Data Records')
df4
```

```python
# count unique OrderID
total_clinics_2020 = df4['OrderID'].nunique()
print(f"Total number of clinics in 2020: {total_clinics_2020}")
```

```python
df4['TotNumCyclesAll']
```

```python
# remove all commas in the "TotNumCyclesAll" column
df4['TotNumCyclesAll'] = df4['TotNumCyclesAll'].str.replace(',', '')

# change the data type of "TotNumCyclesAll" to float
df4['TotNumCyclesAll'] = df4['TotNumCyclesAll'].astype(float)
```

```python
# sum all the cycles in 'TotNumCyclesAll'
total_cycles_2020 = df4['TotNumCyclesAll'].sum()
print(f"National total cycles in 2020: {total_cycles_2020}")
```

```python
# filter to only rows where either "DonorEggService" == Yes or "DonatedEmbryoService"== Yes
df4 = df4[(df4['DonorEggService'] == 'Yes') | (df4['DonatedEmbryoService'] == 'Yes')]
df4
```

```python
# count the unique clinics where patients used donor eggs or embryos in 2020
donor_services_clinics_2020 = df4[(df4['DonorEggService'] == 'Yes') | (df4['DonatedEmbryoService'] == 'Yes')].nunique()
print(f"Clinics offering donor eggs or embryos services in 2020: {donor_services_clinics_2020['OrderID']}")
```

```python
#display every row only with columns  Donor_NumTrans1, Donor_NumTrans2, Donor_NumTrans3, Donor_NumTrans4 
df4_donor = df4[['Donor_NumTrans1', 'Donor_NumTrans2', 'Donor_NumTrans3', 'Donor_NumTrans4']]
```

```python
# check the dtype of the columns
df4_donor.dtypes
```

```python
# change 'Donor_NumTrans1', 'Donor_NumTrans2', 'Donor_NumTrans3', 'Donor_NumTrans4' to float, before that replace all "*" with nan
df4_donor = df4_donor.replace('*', np.nan)
df4_donor = df4_donor.astype(float)
df4_donor.dtypes
```

```python
# Sum Donor_NumTrans1, Donor_NumTrans2, Donor_NumTrans3, Donor_NumTrans4 of each row, and put the number in a new column "TotalDonorTransfers"
df4.loc[:, 'TotalDonorTransfers'] = df4['Donor_NumTrans1']+ df4['Donor_NumTrans2'] + df4['Donor_NumTrans3'] + df4['Donor_NumTrans4']
df4.loc[:, 'TotalDonorTransfers'] = df4['TotalDonorTransfers']
```

```python
print(f"Total number of donor transfers in 2020: {df4['TotalDonorTransfers'].sum()}")
```

```python
# calculate the percentage of cycles in which patients used donor eggs or embryos in 2020
pecentage_of_donor_cycles_2020 = (df4['TotalDonorTransfers'].sum() / total_cycles_2020) * 100
print(f"Percentage of cycles that were donor eggs or embryos transfers in 2020: {pecentage_of_donor_cycles_2020:.2f}% (probably not referencable)")
```

## Join 2010 and 2020 data

```python
df3_select_columns = df3[['PrevClinName1', 'PrevClinName2',	'ClinCityCode',	'ClinStateCode', 'MedicalDirector']]
df4_select_columns = df4[['CurrentClinicName1', 'CurrentClinicName2', 'CurrentClinicCity', 'CurrentClinicState' ,'MedicalDirector']]
merged_on_md = df3_select_columns.merge(df4_select_columns, on='MedicalDirector', how='inner')
merged_on_md
```

```python
# import fuzzywuzzy
from fuzzywuzzy import fuzz

# get fuzzy match score for PrevClinName1 and CurrentClinicName1
merged_on_md['match_score'] = merged_on_md.apply(lambda x: fuzz.ratio(x['PrevClinName1'], x['CurrentClinicName1']), axis=1)
merged_on_md
```

```python
merged_on_md.to_csv('merged_on_md.csv', index=False)
```
