## 1995 data

```python
# read excel "clinic_tables_data_1995.xls" in the folder "Input"
df_1995 = pd.read_excel("Input/clinic_tables_data_1995.xls", sheet_name="1995 Clinic Tables Data", header=0)
df_1995
```

```python
# find the column names: Number_Cycles_LT35, Number_Cycles_35to39, Number_Cycles_GT39, Num_Transfer_LT35, Num_Transfer_35to39, Num_Transfer_GT39, Num_Fresh_TX_LT35, Num_Fresh_TX_35to39, Num_Fresh_TX_GT39
# display the columns, and name the df as df_1995_cycles
df_1995_cycles = df_1995[['Number_Cycles_LT35', 'Number_Cycles_35to39', 'Number_Cycles_GT39', 'Num_Transfer_LT35', 'Num_Transfer_35to39', 'Num_Transfer_GT39', 'Num_Fresh_TX_LT35', 'Num_Fresh_TX_35to39', 'Num_Fresh_TX_GT39']]
df_1995_cycles
```

```python
# rename the columns
# Number_Cycles_LT35 -> fresh_non_<35, Number_Cycles_35to39 -> fresh_non_35-39, Number_Cycles_GT39 -> fresh_non_>39

df_1995_cycles = df_1995_cycles.rename(columns={'Number_Cycles_LT35': 'fresh_non_<35', 'Number_Cycles_35to39': 'fresh_non_35-39', 'Number_Cycles_GT39': 'fresh_non_>39'})
df_1995_cycles
```

```python
# rename: Num_Transfer_LT35 -> frozen_non_<35, Num_Transfer_35to39 -> frozen_non_35-39, Num_Transfer_GT39 -> frozen_non_>39
df_1995_cycles = df_1995_cycles.rename(columns={'Num_Transfer_LT35': 'frozen_non_<35', 'Num_Transfer_35to39': 'frozen_non_35-39', 'Num_Transfer_GT39': 'frozen_non_>39'})
df_1995_cycles
```

```python
# rename: Num_Fresh_TX_LT35	Num_Fresh_TX_35to39	Num_Fresh_TX_GT39
df_1995_cycles = df_1995_cycles.rename(columns={'Num_Fresh_TX_LT35': 'fresh_donor_<35', 'Num_Fresh_TX_35to39': 'fresh_donor_35-39', 'Num_Fresh_TX_GT39': 'fresh_donor_>39'})
df_1995_cycles
```

```python
# turn all data into numeric
df_1995_cycles = df_1995_cycles.apply(pd.to_numeric, errors='coerce')
# leave out all decimals
df_1995_cycles = df_1995_cycles.astype(int)
df_1995_cycles
```

```python
# add the state column from the original df_1995
df_1995_cycles['State'] = df_1995['Clinic_State']
df_1995_cycles
```

### Caculate total cycles & donor cycles


```python
# add a column "total cycles" by summing all numeric columns
df_1995_cycles['total_cycles'] = df_1995_cycles.iloc[:, 0:9].sum(axis=1)
df_1995_cycles
```

```python
# group by state
df_1995_cycles_by_state = df_1995_cycles.groupby('State').sum()
df_1995_cycles_by_state
```

```python
# sum all numbers in "total_cycles" column, and just print the total number of cycles out
total_cycles_1995 = df_1995_cycles_by_state['total_cycles'].sum()
print(f'Total number of cycles in 1995: {total_cycles_1995}')
```

```python
# in df_1995_cycles_by_state, group [fresh_non_<35	fresh_non_35-39	fresh_non_>39	frozen_non_<35	frozen_non_35-39	frozen_non_>39] into a new column "total_cycles_non_donor"
df_1995_cycles_by_state['total_cycles_non_donor'] = df_1995_cycles_by_state['fresh_non_<35'] + df_1995_cycles_by_state['fresh_non_35-39'] + df_1995_cycles_by_state['fresh_non_>39'] + df_1995_cycles_by_state['frozen_non_<35'] + df_1995_cycles_by_state['frozen_non_35-39'] + df_1995_cycles_by_state['frozen_non_>39']
# group [fresh_donor_<35	fresh_donor_35-39	fresh_donor_>39] into a new column "total_cycles_donor"
df_1995_cycles_by_state['total_cycles_donor'] = df_1995_cycles_by_state['fresh_donor_<35'] + df_1995_cycles_by_state['fresh_donor_35-39'] + df_1995_cycles_by_state['fresh_donor_>39']
df_1995_cycles_by_state
```

```python
# only display [ State, total_cycles, total_cycles_non_donor, total_cycles_donor] in the data frame
df_1995_cycles_by_state = df_1995_cycles_by_state[['total_cycles', 'total_cycles_non_donor', 'total_cycles_donor']]
df_1995_cycles_by_state
```

```python
# add a row called "National" to the df_1995_cycles_by_state, which is the sum of all states
df_1995_cycles_by_state.loc['National'] = df_1995_cycles_by_state.sum()
df_1995_cycles_by_state
```

```python
# save this data frame to a csv file in the folder "Output", named "1995_cycles_by_state.csv"
df_1995_cycles_by_state.to_csv("Output/1995_cycles_by_state.csv")
```
